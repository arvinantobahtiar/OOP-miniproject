```python
class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

class Inventory:
    def __init__(self):
        self.products = {}

    def add_product(self):
        name = input("Enter the product name: ").lower()
        price = float(input("Enter the product price: "))
        stock = int(input("Enter the product stock: "))

        product = Product(name, price, stock)
        self.products[name] = product
        print("*Product", name, "successfully added to the inventory.")

    def search_product(self):
        name = input("Enter the product name to search: ").lower()
        if name in self.products:
            print("\n========= SEARCH RESULT ==========")
            product = self.products[name]
            print("Product:", product.name)
            print(f"Price: ${product.price}")
            print("Stock:", product.stock, "pcs")
            print("==================================\n")
        else:
            print("(!)Product not found.")

    def update_stock(self):
        name = input("Enter the product name to update stock: ").lower()
        if name in self.products:
            new_stock = int(input("Enter the new stock: "))
            product = self.products[name]
            product.stock = new_stock
            print("*Stock of", name, "successfully updated to", new_stock, "stock")
        else:
            print("(!)Product not found.")

    def remove_product(self):
        name = input("Enter the product name to remove: ")
        if name in self.products:
            del self.products[name]
            print("*Product", name, "successfully removed from the inventory.")
        else:
            print("(!)Product not found.")

    def display_inventory(self):
        print("\n========== CURRENT INVENTORY ===========")
        for name, product in self.products.items():
            print("Product:", product.name)
            print(f"Price: ${product.price}")
            print("Stock:", product.stock, "pcs")
            print('========================================')

inventory = Inventory()

print("|||||||||||| WELCOME ||||||||||||\n",
        "1. Add Product\n",
        "2. Search Product\n",
        "3. Update Stock\n",
        "4. Remove Product\n",
        "5. Display Current Inventory\n",
        "6. Exit\n",
      "||||||||||||||||||||||||||||||||||\n")

while True :
  method = int(input("Input the number of your option (1-6) : "))
  if method==1 :
    inventory.add_product()
  elif method==2 :
    inventory.search_product()
  elif method==3 :
    inventory.update_stock()
  elif method==4 :
    inventory.remove_product()
  elif method==5 :
    inventory.display_inventory()
  elif method==6 :
    print("Bye Bye")
    break
  else :
      print("input N/A !")

  print()
    
```
